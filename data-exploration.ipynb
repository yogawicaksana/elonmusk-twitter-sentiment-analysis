## Word Count

```python
import numpy as np
from wordcloud import WordCloud
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
from PIL import Image
```

```python
dfLabel.isna().sum()
```

```python
dfLabelClean=dfLabel.dropna()
```

```python
all_words = ''.join([word for word in dfLabelClean['new_tweet']])
wordcloud = WordCloud(stopwords=STOPWORDS, width=800, height=500, random_state=21, max_font_size=150, background_color='white').generate(all_words)
plt.figure(figsize=(15,8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('The words in tweet contain `Elon Musk Twitter` (2022-04-26 22:36:07 -- 2022-04-27 06:59:59)' , weight='bold', fontsize=14)
plt.show()
```

```python
twitter_mask = np.array(Image.open('twitter.png'))

def transform_format(val):
    if val == 0:
        return 255
    else:
        return val

transformed_twitter_mask = np.ndarray((twitter_mask.shape[0], twitter_mask.shape[1]), np.int32)

for i in range(len(twitter_mask)):
    transformed_twitter_mask[i] = list(map(transform_format, twitter_mask[i]))
```

```python
wc = WordCloud(background_color="white", max_words=1000, mask=transformed_twitter_mask,
               stopwords=STOPWORDS, contour_width=3, contour_color='firebrick')

# Generate a wordcloud
wc.generate(all_words)

# store to file
wc.to_file("twitter.png")

# show
plt.figure(figsize=[20,10])
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```

### Word Count Positive Sentiment

```python
dfLabelPositive=dfLabelClean[dfLabelClean['sentiment']==2]
```

```python
all_words = ''.join([word for word in dfLabelPositive['new_tweet']])
wordcloud = WordCloud(width=800, height=500, random_state=21, max_font_size=150, background_color='white').generate(all_words)
plt.figure(figsize=(15,8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Positive sentiment words in tweet contain `Elon Musk Twitter` (2022-04-26 22:36:07 -- 2022-04-27 06:59:59)' , weight='bold', fontsize=14)
plt.show()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import collections

cv=CountVectorizer()
bow=cv.fit_transform(dfLabelPositive['new_tweet'])
word_freq = dict(zip(cv.get_feature_names(),
                 np.asarray(bow.sum(axis=0)).ravel()))
word_counter=collections.Counter(word_freq)
word_counter_df = pd.DataFrame(word_counter.most_common(20), columns=['word', 'freq'])

fig, ax=plt.subplots(figsize=(15,7))
plt.title('Top 20 Most Used Words (Positive Sentiment)', family='Serif', weight='bold', size=20)
sns.barplot(x='word', y='freq', data=word_counter_df, palette='PuBuGn_d', ax=ax)
```

```python
df['date'] = df['date'].apply(pd.to_datetime)
```

```python
# dfGrouped = df.groupby(pd.Grouper(key='date', freq='15min')).count().reset_index()
dfGrouped = df.groupby(pd.Grouper(key='date', freq='15min')).count()
dfGrouped.head(3)
```

#### Merging labeled data and unlabeled data

```python
dfMerged=df.merge(dfLabel, how='inner', on='tweet')
```

```python
dfMerged=dfMerged[dfMerged['new_tweet'].notna()]
```

#### Separate Positive and Negative

```python
dfGroupedPositive = dfMerged[dfMerged['sentiment']==2]
dfGroupedPositive = dfGroupedPositive.groupby(pd.Grouper(key='date', freq='15min')).count()
# dfGroupedPositive = dfGroupedPositive.groupby(pd.Grouper(key='date', freq='15min')).count().reset_index()
dfGroupedPositive.head(3)
```

```python
dfGroupedNegative = dfMerged[dfMerged['sentiment']==0]
dfGroupedNegative = dfGroupedNegative.groupby(pd.Grouper(key='date', freq='15min')).count()
# dfGroupedNegative = dfGroupedNegative.groupby(pd.Grouper(key='date', freq='15min')).count().reset_index()
dfGroupedNegative.head(3)
```
